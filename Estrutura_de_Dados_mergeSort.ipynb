```python
def merge_sort(list_one):
  if len(list_one) <= 1:
    return list_one

  mid = len(list_one) // 2
  left_half = list_one[:mid]
  right_half = list_one[mid:]

  sorted_left = merge_sort(left_half)
  sorted_right = merge_sort(right_half)

  return merge(sorted_left, sorted_right)

def merge(list_left, list_right):
  result = []
  i = j = 0

  while i < len(list_left) and j < len(list_right):
    if list_left[i] < list_right[j]:
      result.append(list_left[i])
      i += 1
    else:
      result.append(list_right[j])
      j += 1

  result.extend(list_left[i:])
  result.extend(list_right[j:])

  return result
```

```python
merge_sort([2, 40, 1, 5, 4])
```
