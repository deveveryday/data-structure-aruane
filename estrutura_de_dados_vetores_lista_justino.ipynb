```python
#21. Segundo Maior Elemento
#Enunciado: Crie uma função que retorne o segundo maior elemento de uma lista. (Considere que a lista pode ter duplicatas.)

def find_second_bigger(the_list):
  greater_number = the_list[0]
  second_greater_number = the_list[1]

  for i in the_list:
    if i > greater_number:
      greater_number = i

  for i in the_list:
    if i > second_greater_number and i < greater_number:
      second_greater_number = i

  return second_greater_number


find_second_bigger([1,2,3,4,5,5])
#----------------------------------------
```

```python
#23. Intersecção de Duas Listas
#Enunciado: Crie uma função que retorne uma lista com os elementos que são comuns a duas listas (sem duplicatas).

def find_common(list_one, list_two):
  list_new = []

  for i in list_one:
    if i in list_two and i not in list_new:
      list_new.append(i)

  return list_new

find_common([1, 3], [1, 2, 3])

#----------------------------------------
```

```python
#26. Duplicar Elementos
#Enunciado: Crie uma função que receba uma lista e retorne uma nova lista onde cada elemento aparece duas vezes consecutivas.

def make_double_list(the_list):
  double_list = []

  for i in the_list:
    double_list.append(i)
    double_list.append(i)

  return double_list


make_double_list([1,2,3])

#----------------------------------------
```

```python
#27. Ordenação sem Sort (Selection Sort)
#Enunciado: Crie uma função que ordene uma lista em ordem crescente implementando o algoritmo de Selection Sort, sem usar métodos nativos de ordenação.

def super_sort(the_list):
  ordered_list = []
  min_index = -1
  min_value = 0

  for i in range(len(the_list)):
    min_index = i
    for j in range(i + 1, len(the_list)):
      if the_list[j] > the_list[min_index]:
        min_index = j

    min_value = the_list.pop(min_index)
    the_list.insert(i, min_value)
    ordered_list.insert(i, min_value)

  return ordered_list


print(super_sort([3, 4, 2, 1]))

#----------------------------------------
```
