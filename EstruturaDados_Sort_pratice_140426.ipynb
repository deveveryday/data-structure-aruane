```python
# Exercício 1 – Ordenação simples
# Dado o vetor v = [5, 3, 8, 1, 2], aplique o algoritmo bubble_sort passo a passo e mostre como o
# vetor muda a cada iteração.

v = [5, 3, 8, 1, 2]

def show_step_bubble_sort(the_list):
    length = len(the_list)

    for i in range(length):
        for j in range(length - 1 - i):
            print(f"{the_list[j]} > {the_list[j + 1]}")
            if the_list[j] > the_list[j + 1]:
                print(the_list[j], the_list[j + 1], the_list[j] > the_list[j + 1])
                the_list[j], the_list[j + 1] = the_list[j + 1], the_list[j]
                print(the_list[j], the_list[j + 1])
                print(the_list)
                
    print(the_list)

show_step_bubble_sort(v)
```

```python
# Exercício 2 – Vetor já ordenado
# Teste o algoritmo com o vetor v = [1, 2, 3, 4, 5].
# O que acontece? Quantas comparações são feitas?
# Objetivo: entender o comportamento do algoritmo no melhor caso.

def count_steps(the_list):
    sum_steps = 0
    length = len(the_list)

    for i in range(length):
        for j in range(length - i - 1):
            sum_steps += 1
            if the_list[j] > the_list[j - i -1]:
                the_list[j], the_list[j + 1] = the_list[j + 1], the_list[j]
    
    return sum_steps

count_steps([1, 2, 3, 4, 5])


import timeit
t = timeit.timeit(lambda: count_steps([1, 2, 3, 4, 5]))
print(t)

```

```python
# Exercício 3 – Vetor em ordem reversa
# Use o bubble_sort para ordenar v = [9, 7, 5, 3, 1].
# Conte quantas trocas foram feitas.
# Objetivo: mostrar o pior caso de desempenho do Bubble Sort.

t = timeit.timeit(lambda: count_steps([1, 2, 3, 4, 5]))
print(t)
t = timeit.timeit(lambda: count_steps([9, 7, 5, 3, 1]))
print(t)
```

```python
# Exercício 4 – Vetor com elementos repetidos
# Ordene v = [4, 2, 4, 1, 3].
# O que acontece com os elementos iguais durante a ordenação?
# Objetivo: reforçar que o algoritmo mantém a estabilidade dos elementos iguais.
v = [4, 2, 4, 1, 3]

show_step_bubble_sort(v)
```
