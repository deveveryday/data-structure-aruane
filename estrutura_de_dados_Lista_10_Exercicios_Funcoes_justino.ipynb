```python
#1. Crie uma função chamada dobro(numero) que receba um número e retorne o dobro dele.

def get_double(num):
  return num * 2

#----------------------------------------
```

```python
#2. Crie uma função chamada area_retangulo(base, altura) que retorne a área de um retângulo.

def calculate_area_rectangle(base, height):
  return base * height

#----------------------------------------
```

```python
#3. Crie uma função chamada media_tres(n1, n2, n3) que retorne a média aritmética de três números.

def mean_three(n1, n2, n3):
  return (n1 + n2 + n3) / 3

#----------------------------------------
```

```python
#4. Crie uma função chamada maior_entre_dois(a, b) que retorne o maior número entre dois valores recebidos.

def find_bigger_each_other(number_a, number_b):
  if number_a > number_b:
    return number_a
  elif number_b > number_a:
    return number_b
  else:
    return None

#----------------------------------------
```

```python
#5. Crie uma função chamada verifica_intervalo(numero) que retorne:
# - "Dentro do intervalo" se o número estiver entre 10 e 50
# - "Fora do intervalo" caso contrário.

def check_interval(number):
  result = "Ok interval" if number > 10 and number < 50 else "Out of interval"

#----------------------------------------
```

```python
#6. Crie uma função chamada contar_vogais(texto) que receba uma string e retorne a quantidade de vogais presentes nela.

def count_vogals(text):
  result = 0

  for t in text:
    result += 1 if t in "AEIOUaeiou" else 0

  return result

#----------------------------------------

count_vogals("SKATEeveryday")
```

```python
#7. Crie uma função chamada soma_lista(lista) que receba uma lista de números e retorne a soma de todos os elementos.

def sum_the_list(my_list):
  result = 0

  for i in my_list:
    result += i

  return result


print(sum_the_list([3, 0]))

#----------------------------------------
```

```python
#8. Crie uma função chamada multiplicar_lista(lista) que receba uma lista de números e retorne o produto (multiplicação) de todos os elementos.

def multiply_list(the_list):
  result = 1

  for i in the_list:
    result *= i

  return result


multiply_list([2, 1, 2])

#----------------------------------------
```

```python
#9. Crie uma função chamada eh_multiplo(a, b) que retorne True se 'a' for múltiplo de 'b' e False caso contrário.

def multiplication_test(a_number, b_number):
  return True if a_number % b_number == 0 else False

#----------------------------------------
```

```python
#10. Crie uma função chamada contador_regressivo(n) que imprima uma contagem regressiva de n até 0 usando um laço de repetição.

def end_counter(time):
  for i in range(time, -1, -1):
    print(i)

end_counter(11)

#----------------------------------------
```
