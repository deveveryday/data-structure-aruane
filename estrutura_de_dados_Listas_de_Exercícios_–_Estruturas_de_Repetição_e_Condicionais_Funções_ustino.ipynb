```python
#6. Leia o preço de um produto e aplique desconto: até 100 (5%), de 100.01 a 300 (10%), acima de 300 (15%). Mostre preço final.

def apply_discount(value, percentual = 10):
  return value - (value * (percentual/100))

product_price = float(input("Tell me the price of Blue Jacket:: $"))
discount = 0.0

if product_price <= 100:
  discount = 5
elif product_price > 100 and product_price <= 300:
  discount = 10
else:
  discount = 15


print(apply_discount(product_price, discount))
#----------------------------------------
```

```python
#7. Leia 3 lados e verifique se formam triângulo. Se formar, classifique: equilátero, isósceles ou escaleno.

side_a = input("Type the size of A side ")
side_b = input("Type the size of B side ")
side_c = input("Type the size of C side ")

if (side_a < side_b + side_c) and (side_b < side_a + side_c) and (side_c < side_a + side_b):
    print("Escaleno")
elif side_a == side_b == side_c:
    print("Equilátero")
elif (side_a == side_b) or (side_a == side_c) or (side_b == side_c):
    print("Isósceles")
#----------------------------------------
```

```python
#18. Crie a função `acima_da_media(notas)` que retorna quantas notas estão acima da média da própria lista.

def above_the_average(score_list):
  result = 0
  mean_score = 0.0
  sum_score = 0

  for i in score_list:
    sum_score += i

  mean_score = sum_score / len(score_list)
  print(mean_score)

  for i in score_list:
    if i >= mean_score:
      result += 1

  return result

above_the_average([5, 5, 5])
#----------------------------------------
```

```python
#20. Leia um número inteiro e informe se ele é primo (use for para testar divisores). Nível 4

def is_prime(number):
  prime = False if number <= 1 else True

  for i in range(2, int(number ** 0.5) + 1):
    if number % i == 0:
      prime = False
      break

  return prime


is_prime(15)

#----------------------------------------
```

```python
#28. Leia um número N e verifique se ele é um número perfeito (soma dos divisores próprios = N).

try_number = int(input("Send the number to test in python class: "))
sum_number = 0

for i in range(1, try_number):
  sum_number += i if try_number % i == 0 else 0

if (try_number == sum_number):
    print("That's right")
else:
    print("Isn't a perfect number")
#----------------------------------------
```
